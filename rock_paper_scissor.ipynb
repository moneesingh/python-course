```python
import random

# Rock
rock = """
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
"""

paper = """
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
"""

scissor = """
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
"""

print("Welcome to rock, paper, scissor game that you are goign to play against the computer!")
choices = [rock, paper, scissor]
choice_indices = [0, 1, 2]

def find_winner(user, computer):
    print(f"You entered: \n{choices[user]}")
    print(f"Computer entered: \n{choices[computer]}")
    if (user == computer):
        print("It's a draw")
        return

    if (user == 1 and computer == 0):
        print("You are the winner")
        return
    elif (user == 2 and computer == 1):
        print("You are the winner")
        return 
    elif (user == 0 and computer == 2):
        print("You are the winner")
        return  
    else:
        print("Computer is the winner")
        return        


user_choice = int(input("Enter your choice(0=rock),(1=paper),(2=scissor): "))
computer_choice = random.choice(choice_indices)
find_winner(user_choice, computer_choice)
```
